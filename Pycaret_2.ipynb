```python
# Importando Bibliotecas

import pandas as pd

# carregando o dataset

data_classification = pd.read_csv('https://raw.githubusercontent.com/lakshay-arora/PyCaret-Example/master/datasets/loan_train_data.csv')
```

#Importando um mÃ³dulo

Dependendo do tipo de problema que vocÃª vai resolver, primeiro vocÃª precisa importar o mÃ³dulo. Na primeira versÃ£o do Pycaret, 6 mÃ³dulos diferentes estÃ£o disponÃ­veis= regressÃ£o, classificaÃ§Ã£o, clustering, processamento de linguagem natural(PLN), detecÃ§Ã£o de anÃ´malias e regra de mineraÃ§Ã£o associada. 
Neste artigo, vamos resolver um problema de classificaÃ§Ã£o e , portanto, vamos importar o mÃ³dulode classificaÃ§Ã£o.


#Inicializando a configuraÃ§Ã£o

O Pycaret executa algumas tarefas bÃ¡sicas de prÃ©- processamento, como ignorar os IDs e as colunas de data, imputar os valores ausentes, codificar as variÃ¡veis categÃ³ricas e dividir o conjunto de dados na divisÃ£o de teste e treino para o restante das etapas de modelagem.

Quando vocÃª executa a funÃ§Ã£o configuraÃ§Ã£o, ela primeiro confirma os tipos de dados e, em seguida, se vocÃª pressionar enter, criarÃ¡ o ambiente para vocÃª prosseguir

```python
# Importando a classificaÃ§Ã£o do Pycaret

from pycaret import classification
```

```python
classification_setup = classification.setup(data_classification, target='Personal Loan') # EmprÃ©stimo Pessoal

```

# Treinando o Modelo


Treinar um modelo no Pycaret Ã© bastante simples. VocÃª sÃ³ precisa usar a funÃ§Ã£o create_model que leva apenas um parÃ¢metro - a abreviaÃ§Ã£o do modelo como uma string.


Aqui vamos primeiro treinar um modelo de Ã¡rvore de decisÃ£o para o qual temos que passar "dt"  e ele retornarÃ¡ uma tabela com pontuaÃ§Ãµes de validaÃ§Ã£o cruzada K-fold de mÃ©tricas de avaliaÃ§Ã£o usadas para a aprendizagem supervisionada:

ClassificaÃ§Ã£o: precisÃ£o, AUC, recall, precisÃ£o F1, Kappa
RegressÃ£o: MAE, MSE, RMSE, R2, RMSLE, MAPE

```python
# Criando o modelo( MÃ¡quina Preditiva)
classification_dt = classification.create_model('dt')
```

```python
# Criando Agrupamento Boosting
boosting = classification.ensemble_model(classification_dt, method= 'Boosting')
```

```python
# Olha que maravilha( Comparando diversas mÃ¡quinas preditivas)

classification.compare_models()
```

```python
# GrÃ¡fico AUC - ROC

classification.plot_model(classification_dt, plot = 'auc')
```

```python
# Curva Precision Recall
classification.plot_model(classification_dt, plot = 'pr')
```

```python
# Validation Curve

classification.plot_model(classification_dt, plot = 'vc')
```

```python
# ImportÃ¢ncia das variÃ¡veis

classification.plot_model(classification_dt, plot = 'feature')
```

```python
# Confusion Matriz
classification.plot_model(classification_dt, plot = 'confusion_matrix')
```

```python
# AvaliaÃ§Ã£o da MÃ¡quinas preditivas
classification.evaluate_model(classification_dt)
```

# Interpretabilidade dos modelos

```python
# Usando a Biblioteca SHAP

classification.interpret_model(classification_dt)
```

```python
# Dados de teste 
test_data_classification = pd.read_csv('https://raw.githubusercontent.com/lakshay-arora/PyCaret-Example/master/datasets/loan_test_data.csv')
```

```python
# Fazendo as novas previsÃµes

predictions = classification.predict_model(classification_dt, data=test_data_classification)
```

```python
# Imprimindo na tela as prediÃ§Ãµes

predictions
```

```python
# Salvando a MÃ¡quina Preditiva Final

classification.save_model(classification_dt, 'decision_tree_1')
```

```python
# Carregando para usar a MÃ¡quina Preditiva.
# Aqui Ã© onde a aplicaÃ§Ã£o, depois de criada, vai trabalhar......

dt_model = classification.load_model(model_name= 'decision_tree_1')
```
